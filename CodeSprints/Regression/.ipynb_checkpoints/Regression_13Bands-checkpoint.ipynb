```python
# Read S2 dataset 
Data1 = pd.read_csv('data/AllBands/S2Data.csv')

# Remove none values
S2Data = Data1.dropna()

# Display the first few rows of our DataFrame
S2Data.head()
print(S2Data)
```

```python
# Getting column names
X_columns = ["2BDA", "3BDA", "NDCI", "B1", "B2", "B3", "B4", "B5", "B6", "B7", "B8", "B8A", "B9", "B10", "B11", "B12"]
Y = MergedData["Chl-a"]
```

```python
#Calculate VIFs on our dataset
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(MergedData[X_columns[0:16]].values, i) for i in range(MergedData[X_columns[0:16]].shape[1])]
vif["features"] = MergedData[X_columns[0:16]].columns
```

```python
vif.round()
```

```python
#OLS model based on Three predictors (2BDA,3BDA,NDCI)
# Extract independent variables (predictors) 
X = MergedData[X_columns]

# Extract dependent variable (prediction)
Y = MergedData['Chl-a']

model = OLS(Y,X)
results = OLS(Y,X).fit()
print(results.summary())
```

```python
# Add model residuals to our dataframe:
MergedData['residual']=OLS(Y,X).fit().resid
MergedData.to_csv('data/AllBands/MergedData.csv')
```

```python
sns.kdeplot(MergedData['residual'].values, shade=True, color='slategrey')
```

```python
# When L2=0, Ridge equals OLS
model = Ridge(alpha=1)
```

```python
# Define model evaluation method
cv = RepeatedKFold(n_splits=16, n_repeats=16, random_state=1)
# Evaluate model
scores = cross_val_score(model, X, Y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# Force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
model.fit(X,Y)
#Print out the model coefficients
print(model.coef_)
```

```python
# When L1=0, Lasso equals OLS
model = Lasso(alpha=0)
```

```python
# Define model evaluation method
cv = RepeatedKFold(n_splits=16, n_repeats=16, random_state=1)
# Evaluate model
scores = cross_val_score(model, X, Y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# Force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
model.fit(X,Y)
#Print out the model coefficients
print(model.coef_)
```

```python
# When L1 approaches infinity, certain coefficients will become exactly zero, and MAE equals the variance of our response variable:
model = Lasso(alpha=10000000)
```

```python
# Define model evaluation method
cv = RepeatedKFold(n_splits=16, n_repeats=16, random_state=1)
# Evaluate model
scores = cross_val_score(model, X, Y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# Force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
model.fit(X,Y)
# Print out the model coefficients
print(model.coef_)
```

```python
# When L1 approaches infinity, certain coefficients will become exactly zero, and MAE equals the variance of our response variable:
model = ElasticNet(alpha=1, l1_ratio=0.2)
```

```python
# Define model evaluation method
cv = RepeatedKFold(n_splits=16, n_repeats=16, random_state=1)
# Evaluate model
scores = cross_val_score(model, X, Y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# Force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
model.fit(X,Y)
#Print out the model coefficients
print(model.coef_)
```
